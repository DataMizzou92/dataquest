### I  will figure out the most expensive vehicles based upon mean price

```python
auto = autos['brand'].value_counts().sort_values().head(20).index
makers = {}

for row in auto:
    brand = autos[autos['brand'] == row]
    price = brand['price'].mean()
    makers[row] = price

print(makers)
```

```python
vehicles = pd.Series(makers)
print(vehicles)
```

```python
df = pd.DataFrame(vehicles, columns=['mean_price'])
pricey = df.sort_values(['mean_price'], ascending=False)
new_df =  pricey[:3]
new_df
```
